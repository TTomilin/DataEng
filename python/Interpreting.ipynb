```python
solar = pd.read_csv('solar_generation.csv')

normalized_solar_columns = ['DE','BE','RO','BG']
x = solar[normalized_solar_columns].values
x_scaled = min_max_scaler.fit_transform(x)
solar_temp = pd.DataFrame(x_scaled, columns=normalized_solar_columns, index = solar.index)
solar[normalized_solar_columns] = solar_temp

solar = solar.set_index('timestamp')
solar_stacked = solar.stack().rename_axis(['timestamp','Country']).rename('data').reset_index()
solar_stacked['timestamp'] = pd.to_datetime(solar_stacked['timestamp'])
solar_stacked = solar_stacked.set_index('timestamp')
```

```python
solar_daily = solar_stacked.groupby(['Country',pd.Grouper(freq='m')]).mean().reset_index()
wind_daily = wind_stacked.groupby(['Country',pd.Grouper(freq='m')]).mean().reset_index()
```

```python
a4_dims = (11.7, 8.27)
fig, ((ax1, ax2))= pyplot.subplots(nrows=2, ncols=1, figsize=a4_dims)

palette1 = sns.color_palette("Oranges_d", 2)
palette2 = sns.color_palette("Blues_d", 2)

Country_sub1 = wind_daily[wind_daily['Country'].isin(normalized_wind_columns[:2])]
Country_sub2 = wind_daily[wind_daily['Country'].isin(normalized_wind_columns[2:])]
ax1 = sns.lineplot(ax=ax1, data=Country_sub1,x='timestamp',y='data', hue='Country', palette=palette1)
ax2 = sns.lineplot(ax=ax2, data=Country_sub2,x='timestamp',y='data', hue='Country', palette=palette2)

legend = ax1.legend(loc='best', shadow=True, fontsize='large', labels=('Lithuania','Latvia'))
legend = ax2.legend(loc='best', shadow=True, fontsize='large', labels=('Austria','Hungary'))

ax1.set(xlabel='', ylabel='Generation (MWh)',title='Normalized Monthly Wind Energy Generation')
ax2.set(xlabel='Date', ylabel='Generation (MWh)')
fig.savefig("lineplot_wind_correlated_pairs.png", dpi=300)
```
